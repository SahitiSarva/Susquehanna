```python
payoffs = list(range(1,100))
utilities = []
fig = plt.figure(figsize = (6,4))
ax = fig.subplots(nrows = 1, ncols = 1, sharex = True, sharey = True )

for rho in np.arange(0,3.5,0.5).tolist():
    plt.plot(normalize(get_marginal_utility_of_df(pd.DataFrame(payoffs), rho)), color = cm.viridis(rho/2.5), label = rho)
    utilities = get_marginal_utility_of_df(pd.DataFrame(payoffs), rho)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.set_ylabel('utility', font = 'Georgia', fontsize = 12)
    ax.set_xlabel('outcome', font = 'Georgia', fontsize = 12)

    ax.tick_params(axis=u'both', which=u'both',length=0, labelbottom = True)

    plt.legend(loc = 'center right', bbox_to_anchor= (1.15, 0.3), fontsize = 12, frameon = False, title = 'marginal utility of consumption (rho)', prop = 'Georgia', )
    #fig.show()
    plt.savefig('../output/processed/dimishing_utility.png')
```

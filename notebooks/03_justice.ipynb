```python
%reload_ext autoreload
%autoreload 2
from utilities import *
import pickle
import pandas as pd 
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
utility_function = get_marginal_utility_of_df
payoffs = load_data('../model/data/processed/50yr_all2.pickle')
fairness = get_fairness_metrics_with_baseline('Status Quo', utility_function,payoffs )
```

```python
fallback = pd.read_csv('../output/processed/fallback_bargaining.csv')
fallback = fallback[['decision', 'baseline', 'resultant', 'least_preferred']]
fallback = fallback[fallback['baseline']=='Status Quo']
```

```python
fallback = fallback[fallback['least_preferred']==0]
```

```python
noncoop = pd.read_excel('../output/processed/stable_decisions.xlsx')
noncoop = noncoop[['decision', 'stability', 'baseline', 'nr_actors']]
noncoop = noncoop[noncoop['baseline']=='status_quo']
```

```python
value_comparison = fairness.merge(noncoop.merge(fallback, on = 'decision'), on = 'decision')
value_comparison['gini_index'] = value_comparison['gini_index']*-1

value_comparison = value_comparison[['decision', 'gini_index', 'utility', 'stability', 'resultant', 'flood_risk']]
```

```python
min(value_comparison['flood_risk'])
```

```python
def utility_percentile(utility, all_utility):
    if utility <= np.percentile(all_utility, 90):
        return 0
    else:
        return 1
```

```python
value_comparison['utilitarian_percentile'] = value_comparison['utility'].apply(lambda x: utility_percentile(x, value_comparison['utility']) )
value_comparison['egalitarian_percentile'] = value_comparison['gini_index'].apply(lambda x: utility_percentile(x, value_comparison['gini_index']) )
```

```python
import plotly.express as px
df = value_comparison
fig = px.parallel_coordinates(df, color = 'utilitarian_percentile', dimensions = ['utility','gini_index', 'flood_risk','stability', 'resultant'], color_continuous_scale=['lightgrey', '#450256'] , 
labels = {'flood_risk':'Prioritarian towards flood risk','utility': 'Utilitarian', 'gini_index' : 'Egalitarian', 'resultant' : 'Cooperative stability','stability' : 'Non cooperative stability'})

# Removes the colour legend
fig.update_coloraxes(showscale = False)

fig.update_layout(width = 1000, height = 400, font_family = "Georgia", font_size = 12, 
 title = {'text': "Moral tradeoffs among Pareto optimal policies", 'xanchor': 'center', 'yanchor': 'top', 'x': 0.5, 'y': 0.1}
)
fig.show()
fig.write_image('../visuals/moral_tradeoffs.png')
```
